```python
#q9
import re
def extract_damage_level(damage_str):
        # Extract numerical part from the string
        match = re.search(r'\d+', damage_str)
        if match:
            return int(match.group())
        else:
            return 0

df=pd.merge(damages,units, on='CRASH_ID', how='inner')
fil_df=df[(df['FIN_RESP_TYPE_ID']=='PROOF OF LIABILITY INSURANCE') & (~df['VEH_DMAG_SCL_1_ID'].isin(['NA','NO DAMAGE'])) 
          & (~df['VEH_DMAG_SCL_2_ID'].isin(['NA','NO DAMAGE']))]
fil_df=fil_df[fil_df['DAMAGED_PROPERTY'].isin(['NONE','NONE1'])]

fil_df['VEH_DMAG_SCL_1_VAL']= fil_df['VEH_DMAG_SCL_1_ID'].astype(str).apply(extract_damage_level)
fil_df['VEH_DMAG_SCL_2_VAL']= fil_df['VEH_DMAG_SCL_2_ID'].astype(str).apply(extract_damage_level)

fil_df[(fil_df['VEH_DMAG_SCL_1_VAL'] > 4) | (fil_df['VEH_DMAG_SCL_2_VAL']>4)]['CRASH_ID'].nunique()
```

```python
#q10

top_10_colors= units['VEH_COLOR_ID'].value_counts().head(10).index
top_25_states=units['VEH_LIC_STATE_ID'].value_counts().head(25).index

mer_df1=pd.merge(charges, primary_person, on='CRASH_ID', how='inner')
mer_df2=pd.merge(mer_df1, units, on='CRASH_ID', how='inner')
mer_df2['CHARGE'] = mer_df2['CHARGE'].fillna('')
df= mer_df2[(mer_df2['CHARGE'].str.contains('SPEED')) & (mer_df2['DRVR_LIC_TYPE_ID'].isin(['DRIVER LICENSE','COMMERCIAL DRIVER LIC.']))]
color_state_df= df[(df['VEH_COLOR_ID'].isin(top_10_colors)) & (df['VEH_LIC_STATE_ID'].isin(top_25_states))]

color_state_df['VEH_MAKE_ID'].value_counts().head(5).index
```
